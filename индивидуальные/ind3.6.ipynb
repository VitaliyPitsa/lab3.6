задача из области математики - построение трехмерного графика функции двух переменных $f(x, y) = x^2 + y^2$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

 мы можем определить функцию, которую мы будем анализировать. Для примера, мы определим функцию $f(x, y) = x^2 + y^2$.

```python
def f(x, y):
    return x**2 + y**2
```

```python
# создаем матрицу с координатами x и y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
# вычисляем значения функции f для каждой пары координат
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap=plt.cm.jet, rstride=1, cstride=1, alpha=0.7)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('График функции $f(x, y) = x^2 + y^2$')
plt.show()
```
