#### Задание 
Создать ноутбук, в котором выполнить решение вычислительной задачи (например, задачи из области физики, экономики, математики, статистики и т. д.) требующей построения трехмерного графика.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

затем мы можем определить функцию, которая будет создавать модель бублика. Для этого мы можем использовать параметрическое уравнение, которое задает координаты x, y и z модели.

$\begin{equation*}
 \begin{cases}
x = (R_1+R_2\cos(t))\cos(\pi)\\
y = (R_1+R_2\cos(t))\cos(\pi)\\
z = R_2\sin(t) \end{cases}
\end{equation*}$

```python
def bublik(R1, R2, N):
    theta = np.linspace(0, 2*np.pi, N)
    phi = np.linspace(0, 2*np.pi, N)
    theta, phi = np.meshgrid(theta, phi)
    x = (R1 + R2*np.cos(theta)) * np.cos(phi)
    y = (R1 + R2*np.cos(theta)) * np.sin(phi)
    z = R2 * np.sin(theta)
    return x, y, z
```

```python
# создываем данные для бублика
R1 = 0.5
R2 = 0.3
N = 100
x, y, z = bublik(R1, R2, N)
```

```python
fig = plt.figure(figsize=(7,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Модель бублика')
plt.show()
```
